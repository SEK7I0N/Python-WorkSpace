### Univariate analysis 
It is the simplest form of data analysis where the data being analyzed contains only one variable. Since it's a single variable it doesn't deal with causes or relationships. The main purpose of univariate analysis is to describe the data and find patterns that exist within it.

### Bivariate analysis 
It is one of the simplest forms of quantitative (statistical) analysis. It involves the analysis of two variables (often denoted as X, Y), for the purpose of determining the empirical relationship between them. Bivariate analysis can be helpful in testing simple hypotheses of association.

### Multivariate analysis (MVA) 
It is based on the principles of multivariate statistics, which involves observation and analysis of more than one statistical outcome variable at a time. Typically, MVA is used to address the situations where multiple measurements are made on each experimental unit and the relations among these measurements and their structures are important.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#we can use sns.load_dataset(DATASET_NAME) to load in-built example dataSets.
df = sns.load_dataset('tips')
```

```python
# In this DATASET tip is the dependent feature and the rest features like 
#total_bill, sex, size, etc. are independent features in this DATASET 

df.head()
```

```python
#Here we have ploted a graphs using matplotlib between 2 features 
#tip and total_bill
plt.plot(df['tip'],df['total_bill'],'g*')
plt.xlabel('tip')
plt.ylabel('total_bill')
plt.show()

#tip and size
plt.plot(df['tip'],df['size'],'r*')
plt.xlabel('tip')
plt.ylabel('size')
plt.show()

#size and total_bill
plt.plot(df['size'],df['total_bill'],'b*')
plt.xlabel('size')
plt.ylabel('total_bill')
plt.show()
```

### Correlation
A mutual relationship or connection between two or more things.

### Correlation with Heatmap
A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. The values of the first dimensions appear as rows of the table, while the values of the second dimension are represented by the columns of the table.

```python
df.iloc[0:32:8,3]
df.loc[0:32:8]
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

### Join Plot
It can be used for Univariate and Bivariate analysis

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex',color='r')
```

### Pair Plot
It shows all possible Bivariate graph which can be ploted in the provide DATASET.

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='time')
```

### Dist Plot
Dist plot helps us to check the distribution of the columns features.

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False)
```

```python
sns.displot(df['tip'])
```

## Categorical Plot
Seaborn helps us in doing analysis on Categorical Data points.
- Box Plot
- Violin Plot
- Count Plot
- Bar Plot

### Count Plot

```python
sns.countplot(x='day', data=df)
```

### Bar Plot

```python
sns.barplot(x='tip', y='sex', data=df)
```

### Box Plot

```python
sns.boxplot(x='day', y='total_bill', data=df)
```

### Boxen Plot

```python
sns.boxenplot(x='sex', y='total_bill',hue='smoker', data=df)
```

### Violin Plot

```python
sns.violinplot(x='day', y='total_bill', data=df)
```
