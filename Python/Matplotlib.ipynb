# Matplotlib

We need to use following statement to import Matplotlib module into our environment.

```python
import matplotlib.pyplot as plt

#to show the graphs in line without using plt.show() we need this line
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

## Ploting using Matplotlib
### 1) Scatter Plot => plt.scatter()

```python
plt.scatter(x,y,c='b')

#labeling the graph
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph in 2D')

#saving the ploted Graph
plt.savefig('testPlt.png')
```

### 2) Line plot => plt.plot()

```python
a = np.linspace(1,21,27)
b=a**2
```

```python
plt.plot(a,b,'r*-')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph in 2D')
```

```python
#Creating Subplots

plt.subplot(2,2,1)
plt.plot(a,b,'r*-')

plt.subplot(2,2,4)
plt.plot(b,a*np.pi,'go--')

plt.subplot(2,2,2)
plt.plot(a,b**3,'b--')

plt.subplot(2,2,3)
plt.plot(a,b,'g')
```

```python
#Creating Subplots

plt.subplot(2,2,1)
plt.plot(a,np.sin(b),'r*-')

plt.subplot(2,2,4)
plt.plot(np.sin(np.cos(a)),np.sin(a),'go--')

plt.subplot(2,2,2)
plt.plot(a,np.sin(np.cos(b)),'b--')

plt.subplot(2,2,3)
plt.plot(a,np.cos(b),'g')
```

### 3) Bar plot => plt.bar()

```python
plt.bar(a*2,abs(np.cos(b)-np.sin(a)))
plt.bar(a*2,abs(np.cos(b)+np.sin(a)),color = 'r')
```

### 4) Histogram plot => plt.hist()

```python
plt.hist(a)
```

### 5) Box plot => plt.plot()

```python
data  = [np.random.normal(0 ,std, 100) for std in range(1,6)]
```

```python
plt.boxplot(data, vert=True,patch_artist=True)
plt.show()
```
